```python
##NEURAL NETWORKS

##CNN  (CONVOLUTIONAL NEURAL NETWORKS )
##a Convolutional Neural Network or CNN is a type of artificial neural network, which is widely used for image/object recognition and classification.
##Deep Learning thus recognizes objects in an image by using a CNN.

import tensorflow as tf

```

```python
mnist = tf.keras.datasets.mnist
```

```python
mnist
```

```python
(x_train,y_train),(x_test,y_test) = mnist.load_data()
x_train,x_test = x_train/255.0, x_test/255.0
```

```python
mnist = tf.keras.datasets.mnist
## mnist = Modified National Institute of Standards and Technology database
## it has a collection of greycall images with handwritten digits from 0 to 9
```

```python
## unwanted materials are going to deleted automatically by using this code

model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),
    tf.keras.layers.Dense(128,activation = 'relu'),
    tf.keras.layers.Dropout(0,2),
    tf.keras.layers.Dense(10,activation = 'softmax')## multiclass classification purpose  by talking 10 nodes as input
])
```

```python
## x_train,x_train = x_train/255.0,x_test/255.0
## 1--> why to divide by 255
## when we are working with image data,the pixel values are intergers in the range of [0,255]. so, dividing it by 255.0 scales these values to the range of [0,1]
## working with the smaller values increase the stability of optimization algorithm

## 2---> tf.keras.layers.Dense(128,activation = 'relu')
## why 128
## it is the specific number of neurons or units in the dense layer
## relu--> Rectified Linear Unit , it helps the add
## non-linearity to our algorithm
```

```python
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy',metrics = ['accuracy'])
# inorder to avoid loss of info we use this statement
```

```python
model.fit(x_train, y_train, epochs = 5) ## epochs --> how many algorithms have to work at  different train values
```

```python
test_loss,test_accuracy = model.evaluate(x_test,y_test)
print(test_loss)
print(test_accuracy)
```

```python
## *****************Activation function*********************
## RELU-->
## it is one if the widely used activation function , it replaces all negative values with 0 and positive values are kept unchanged

##Sigmoid-->
## it reduces the output between 0 and 1, making it suitable for binary classification problem

## tanh (HYPERbOLIC TANGET)-->
## it reduces the output between -1 to 1

## softmax-->
## it is mostly used in output layer

## Leaku Relu -->
## it is another varient of relu function that allows a small positive gradient for negative value just to avoid dead neurons

```

```python
## PROJECT : 1
# convert image to grey scale using CNN

import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import matplotlib.pyplot as plt
import numpy as np
```

```python
# load the RGB(red,green,blue) image
```

```python
image_path = "/content/download.jfif"
original_image = load_img(image_path,target_size = (224, 224))
original_array = img_to_array(original_image)
original_array = original_array/255.0
print(original_array)
```

```python
original_array.shape
```

```python
plt.figure(figsize = (10,20))
plt.subplot(1,2,1)
plt.title("Original image")
plt.imshow(original_array)
```

```python
# convert image to grey scale

model = models.Sequential()
```

```python
# sequential()--> it allows us to create a time linear stack of layers have connections only to the previous and next layer

model.add(layers.Conv2D(1,(3,3),activation = 'relu',input_shape = (224,224,3))) ## inorder to create a 2d layer
```

```python
# cov2d---> it is represent 2d conventional array

## conventinal layer--> used to perform element wise
#  multiplication or addition or division etc....

# layers.conv2d(1)-->1 is used to mention the number of filters in the comventional layer
# (3,3)--> set the size of the filter

# (224,224,3)--> 224x224-->pixels with 3 color channel(RGB)

model.add(layers.MaxPooling2D(2,2))

# maxpool2d()--> it is used to get the dimensions of input data and extract important fetaures from conv2d layer
```

```python
model.summary()
```

```python
# to reshape the image

input_image = np.expand_dims(original_array,axis = 0)
greyscale = model.predict(input_image)

plt.figure(figsize = (6,6))
plt.subplot(1,2,1)
plt.title("Original image")
plt.imshow(original_array)

plt.figure(figsize = (6,6))
plt.subplot(1,2,1)
plt.title(" Grey scale image(modified)")
plt.imshow(np.squeeze(greyscale),cmap = 'gray')
```

```python
def assign_label(img, flower_type):
  return flower_type
```

```python
## tqdm ---> it creates a progress bar from loop
def make_train_data(flower_type, DIR):
  for img  in tqdm(os.listdir(DIR)):
    label = assign_label(img, flower_type)
    path = os.path.join(DIR, img)
    img = cv2.imread(path, cv2.IMREAD_COLOR)
    img = cv2.resize(img, (IMG_SIZE,IMG_SIZE))

    x.append(np.array(img))
    z.append(str(label))
```

```python
## LINEAR REGRESSION

## PROJECT - 1

## Importing pandas
# loading the file "Linear_regr_Salary_dataset.csv"

import pandas as pd
data = pd.read_csv("/content/Linear_regr_Salary_dataset.csv")
data
```

```python
## Printing the size of the csv file

data.shape
```

```python
## checking  there are any null values in the dataframe Or not

data.isnull().sum()
```

```python
## storing values in the x and y variables

x = data[['YearsExperience']]
y = data[['Salary']]
```

```python
# Training the content

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.4, random_state=58)
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model
```

```python
## Fitting the model that we created

model.fit(x_train,y_train)
```

```python
# Trainded x variable values

x_train
```

```python
# Predicting the x testing value and stroring

y_pred = model.predict(x_test)
y_pred
```

```python
## Printing the size of the y_pred variable
y_pred.shape
```

```python
## Printing the column values of the y_test
y_test
```

```python
## Finding the accuarcy of the trained data

# importing numpy module

import numpy as np
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,np.round(y_pred))
accuracy

```

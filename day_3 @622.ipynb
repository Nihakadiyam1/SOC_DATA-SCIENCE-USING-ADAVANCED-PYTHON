```python
#to to convert dictionary to dataframe
d = {'A':[1,2,np.nan],"B":[5,np.nan,np.nan],"C":[1,2,3],"D":[np.nan,np.nan,np.nan]}
df1 = pd.DataFrame(d)
df1
```

```python
#to drop the value with nan
#it works with columns and rows to delete
#dropna()
df1.dropna()
```

```python
#dropna(how,thresh,axis,inplace)
#how="all" or "any"

df1.dropna(how = "all") # it will be checking the rows with all Nan values to delete that row(-->toget deleted all rows must contain nan values)
```

```python
df1.dropna(how = "any") #it will be deleting the entire row having any one of Nan values in any of the row value
```

```python
df1.dropna(how = 'all',axis = 1)# deletes the column if all the column values are Nan values
```

```python
#to check the sum of nan value in each column
#prints true if it contains Nan values
df1.isna()
```

```python
df1.isna().sum()
```

```python
df1
```

```python
#thresh --> t0 delete the rows
df1.dropna(thresh=3)# here in value place--> if it is 1 --> it must contain any one " non Nan value"
```

```python
df1.dropna(thresh=1)
```

```python
df1.dropna(thresh=2,axis=0)
```

```python
#fillna()--> here we dont want to worry about nan values
df1.fillna(value=0)
```

```python
#to fill the nan values with the average of remainig values in column
df1['A'].fillna(value = df1['A'].mean())
```

```python
#to plot many objects in graph
years = np.arange(2000,2005,dtype=int)
innova=np.array([20,40,10,14,23])
honda = np.array([10,15,20,25,30])
sales = np.array([23,29,37,49,59])
cost = np.array([11,22,33,44,55])
years,innova
plt.plot(years,innova,"--+g")
plt.plot(years,honda,"--+r")
plt.plot(years,sales,"--+y")
plt.plot(years,cost,"--+p")
plt.xlabel("Years")
plt.ylabel("No of years")
plt.title("Car Statistics")
plt.legend(['Innova','City'])

```

```python
#to print the bargraph of male and female in a titanic ship from the titanic file
gender1 = (df5['Sex']== 'female').sum()
gender2 = (df5['Sex']== 'male').sum()
plt.bar(['female','male'],[gender1,gender2],width = 0.5,color=['violet','yellow'])
```

```python
# Filter the dataset to include only the passengers who died
died_passengers = df5[df5["Survived"] == 0]

# Count the number of male and female passengers who died
gender_counts = died_passengers["Sex"].value_counts()

# Plot a bar graph
plt.bar(gender_counts.index, gender_counts.values, color=['skyblue', 'pink'],width = 0.5)
plt.title('Number of Male and Female Passengers Who Died on the Titanic')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

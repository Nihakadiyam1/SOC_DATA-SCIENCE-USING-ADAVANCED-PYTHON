```python
#--------> unsupervised learning algorithms
##  k means clustering

## k means clustering is an unsupervised learning algorithm that will attempt tp group similar clusters together from your data

## it is mainly used in
# cluserting similar documents
#clustering customers based on similar featutres

```

```python
from sklearn.datasets import make_blobs

## make_blobs is used to generate synthetic datasets for clusterung and classification tasks

# this function will create clusters of data points with gaussian distribution
```

```python
##creating random dataset

data = make_blobs(n_samples = 200,n_features = 2,centers = 5,cluster_std = 1.6,random_state=101)
data
```

```python
data = make_blobs(n_samples = 200,n_features = 2,centers = 4,cluster_std = 1.6,random_state=101)
# n_samples = Total number of points equally divided among clusters.
# n_features = It indicates the number of features(columns)
# centers = It determines number of clusters to be generated
# clusters_std = It sets the standard deviation of the clusers, High value makes the clusters to spread out
data
```

```python
## scatter plot

import matplotlib.pyplot as plt

x,y = data

plt.scatter(x[:,0],x[:,1],c = y)
plt.xlabel("Features 1")
plt.ylabel("Features 2")
plt.title("Scatter plot for k means")
```

```python
## size of the dataframe
data[0].shape
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
kmeans.fit(data[0])
kmeans.cluster_centers_
```

```python
## plotiings of kmeans and actual graphs

fig, (ax1,ax2) = plt.subplots(1,2,figsize=(10,6))
ax1.set_title('K means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')
ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df_feat

y = df['TARGET CLASS']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3)

x.shape
```

```python
x_train.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
##based on the neighbors we can relate to the plottings
KNN = KNeighborsClassifier(n_neighbors = 3)
```

```python
## to train model

KNN.fit(x_train, y_train)
```

```python
pred = KNN.predict(x_test)
pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(pred,y_test)
```

```python
acc
```

```python
# to find the error rate

error_rate = []
for val in range(1,30):
  knn = KNeighborsClassifier(n_neighbors=val)
  knn.fit(x_train,y_train)
  pred_i = knn.predict(x_test)
  error_rate.append(np.mean(pred_i != y_test))
```

```python
error_rate
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,3))
plt.plot(range(1,30),error_rate)
plt.title("Error rate vs value")
plt.xlabel("K Value")
plt.ylabel("Error Rate")
```
